```python
#date range

#download aapl stock price historical data from nasdaq

import pandas as pd
df = pd.read_csv("aapl_no_dates.csv")
df.head()
```

```python
df.info()
```

```python
#selecting stock prices for a certain date range -- only business days excluding weekends

rng = pd.date_range(start="2017-06-01", end="2017-06-30", freq='B')
rng
```

```python
df.set_index(rng, inplace=True)
df
```

```python
df.Close.mean()
```

```python
%matplotlib inline
df.Close.plot()
```

```python
#we want to insert weekend days also and keep stock prices as friday's prices.. method = pad, will take the last row values to missing date rows

df.asfreq('D', method='pad')
```

```python
#we can also use padding method Weekly

df.asfreq('W', method='pad')
```

```python
#we can also use padding method to show Hourly

df.asfreq('H', method='pad')
```

```python
#if you dont know your end dates and only the start date and number of periods

rng = pd.date_range(start='2017-06-01', periods=72, freq='B')
rng
```

```python
#or hourly freq, 72 periods

rng = pd.date_range(start='2017-06-01', periods=72, freq='H')
rng
```
